```python
page = requests.get('https://www.telemart.pk/men-s-fashion/footwear/formal.html')
soup = BeautifulSoup(page.content,'html.parser')
```

```python
print(main_class_data.prettify())
```

# Now Find All Products Data

```python
Discounts = [Ds.get_text() for Ds in main_class_data.select('.sale-item .first')]
```

```python
for i in range(9):
    Discounts.append(np.nan)
```

```python
Discounts
```

```python
productNames = [pN.get_text() for pN in main_class_data.select('.pro-inner .pro-title')]
```

```python
productNames
```

```python
currentPrices = [cP.get_text()[1:] for cP in main_class_data.select('.special-price .price')]
```

```python
CC = [cc.get_text() for cc in main_class_data.select('.regular-price .price')]
currentPrices.extend(CC)
```

```python
currentPrices
```

```python
Regular_old_prices = [old.get_text()[1:] for old in main_class_data.select('.old-price .price')]
```

```python
for i in range(9):
    Regular_old_prices.append(np.nan)
```

```python
Regular_old_prices
```

```python
stockStatus = [stockS.get_text() for stockS in main_class_data.select('.wrap-pio .ins')]
```

```python
stockStatus
```

```python
ImageSrc = [img['data-src'] for img in main_class_data.select('.product-image img')]
```

```python
ImageSrc
```

#  Display All extracted Data

```python
pd.DataFrame({
    'productName':productNames,
    'currentPrice':currentPrices,
    'oldPrice':Regular_old_prices,
    'Discount':Discounts,
    'stockCheck':stockStatus,
    'Image':ImageSrc
})
```
